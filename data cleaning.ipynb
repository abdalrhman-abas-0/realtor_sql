> dropping the environmental risk & monthly payment columns as they show case no relevant information & changing the neighborhood column to reflect the neighborhood in which the house is located as this will be handy when creating an sql data base to group the houses by neighborhood.

```python
os.listdir('input files')
```

```python
f_list = os.listdir('input files')
for file in f_list:

    df = pd.read_json(input_dir + file)
    c_df = df.drop_duplicates(subset = ['listing page'])
    
    if "environmental risk" in c_df.columns:
        c_df.drop(columns= ["environmental risk"], inplace= True)
        
    if "monthly payment" in c_df.columns:
        c_df.drop(columns= ["monthly payment"], inplace= True)
        
    neighborhood_name = re.search(r"(?<=in\s{1})\w+(?=-)",file)[0].replace("_"," ")
    c_df["neighborhood"] = neighborhood_name
            
    c_df.to_json(output_dir + file)
    print(file, f"<<<<{len(c_df)}>>>>")
```

```python
# Building a function that turns the school column value from str to list  and adds 
# the house listing page to each school in the school list
schools_all = []
def school_f(schools,listing_page):
    s_df = pd.DataFrame(list(schools))
    s_df["listing page"] = listing_page
    
    global schools_all
    if len(schools_all) == 0:
        schools_all = pd.DataFrame(columns = list(s_df.columns))
    schools_all = pd.concat([schools_all,s_df], ignore_index = True)
```

```python
# Building a function that turns the price_history column value from str to list  and adds 
# the house listing page to each price_history in the price_history list
price_history_all = {}
def price_history_f(ph_df, listing_page):
    ph_df = pd.DataFrame(json.loads(ph_df))
    ph_df.drop(columns= ['profile'], inplace= True)
    ph_df["listing page"] = listing_page
    global price_history_all
    if len(price_history_all) == 0:
        price_history_all = pd.DataFrame(columns = list(ph_df.columns))
    price_history_all = pd.concat([price_history_all, ph_df], ignore_index = True)

```
